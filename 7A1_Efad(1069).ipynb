```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats


scores = np.array([93,83,69,62,75,93,73,77,65,65,78,90,94,78,57,76,56,78,84,92])

mean_score = np.mean(scores)
pop_variance = np.var(scores, ddof=0)   # population variance
sample_variance = np.var(scores, ddof=1) # sample variance
pop_std = np.std(scores, ddof=0)
sample_std = np.std(scores, ddof=1)

print("Mean:", mean_score)
print("Population Variance:", pop_variance)
print("Sample Variance:", sample_variance)
print("Population Std Dev:", pop_std)
print("Sample Std Dev:", sample_std)


np.random.seed(42)
generated_scores = np.random.normal(mean_score, pop_std, 20)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(scores, bins=8)
plt.title("Real Exam Scores")
plt.xlabel("Score")
plt.ylabel("Count")

plt.subplot(1,2,2)
plt.hist(generated_scores, bins=8)
plt.title("Generated Scores (Normal Dist.)")
plt.xlabel("Score")
plt.ylabel("Count")

plt.tight_layout()
plt.show()


mc_samples = np.random.normal(mean_score, pop_std, 100000)
prob_75_or_more = np.mean(mc_samples >= 75)
print("Probability Score â‰¥ 75:", prob_75_or_more)


mu0 = 69  # Null hypothesis mean
t_stat, p_value = stats.ttest_1samp(scores, mu0)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Visualization of Hypothesis
plt.figure(figsize=(6,4))
plt.hist(scores, bins=8)
plt.axvline(mean_score, linestyle='dashed', linewidth=2, label=f"Sample Mean ({mean_score:.2f})")
plt.axvline(mu0, linestyle='solid', linewidth=2, label=f"H0 Mean ({mu0})")
plt.title("Hypothesis Test Visualization")
plt.xlabel("Exam Scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
