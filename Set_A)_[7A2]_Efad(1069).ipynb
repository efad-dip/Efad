```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats


arrivals = np.array([8,12,10,15,9,13,11,14,7,16,10,12,9,15,11,13,14,8,12,10])


mean_arr = np.mean(arrivals)
pop_var = np.var(arrivals, ddof=0)
sample_var = np.var(arrivals, ddof=1)
std_arr = np.std(arrivals, ddof=1)

print("Mean:", mean_arr)
print("Population Variance:", pop_var)
print("Sample Variance:", sample_var)
print("Standard Deviation:", std_arr)


np.random.seed(42)
simulated_arrivals = np.random.uniform(low=arrivals.min(), high=arrivals.max(), size=30)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(arrivals, bins=8)
plt.title("Real Customer Arrivals")
plt.xlabel("Daily Arrivals"); plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(simulated_arrivals, bins=8)
plt.title("Simulated (Uniform) Arrivals")
plt.xlabel("Daily Arrivals"); plt.ylabel("Frequency")

plt.tight_layout()
plt.show()


mc_samples = np.random.uniform(low=arrivals.min(), high=arrivals.max(), size=100000)
prob_over_15 = np.mean(mc_samples > 15)
print("Probability of more than 15 arrivals:", prob_over_15)


hyp_mean = 19
t_stat, p_value = stats.ttest_1samp(arrivals, hyp_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)


plt.figure(figsize=(6,4))
plt.hist(arrivals, bins=8)
plt.axvline(mean_arr, linestyle='dashed', linewidth=2, label=f"Sample Mean ({mean_arr:.2f})")
plt.axvline(hyp_mean, linestyle='solid', linewidth=2, label=f"Hypothesized Mean ({hyp_mean})")
plt.title("Sample Mean vs Hypothesized Mean")
plt.xlabel("Daily Arrivals"); plt.ylabel("Frequency")
plt.legend()
plt.show()
```
